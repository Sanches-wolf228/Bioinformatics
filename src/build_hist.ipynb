```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_1 = pd.read_csv('H3F3A_bipolar_neuron.ENCFF068SXT.hg38.bed', header = None, sep = '\t')
df_2 = pd.read_csv('H3F3A_bipolar_neuron.ENCFF959PEX.hg38.bed', header = None, sep = '\t')
df_1['len'] = df_1[2] - df_1[1]
df_2['len'] = df_2[2] - df_2[1]
```

```python
plt.figure(figsize = (20, 8))
plt.rcParams['font.size'] = 18

plt.subplot(1, 2, 1)
sns.distplot(df_1['len'], color = 'blue', kde = False)
plt.title("H3F3A_bipolar_neuron.ENCFF068SXT.hg38.bed\n" + "Number of peaks =" + str(df_1.shape[0]))
plt.xlabel("Len")
plt.ylabel("Count")

plt.subplot(1, 2, 2)
sns.distplot(df_2['len'], color = 'red', kde = False)
plt.title("H3F3A_bipolar_neuron.ENCFF959PEX.hg38.bed\n" + "Number of peaks =" + str(df_2.shape[0]))
plt.xlabel("Len")
plt.ylabel("Count")

plt.savefig("hg38.histplot.png", format="png")

plt.show()
```

```python
df_3 = pd.read_csv('H3F3A_bipolar_neuron.ENCFF068SXT.hg19.bed', header = None, sep = '\t')
df_4 = pd.read_csv('H3F3A_bipolar_neuron.ENCFF959PEX.hg19.bed', header = None, sep = '\t')
df_3['len'] = df_3[2] - df_3[1]
df_4['len'] = df_4[2] - df_4[1]
```

```python
plt.figure(figsize = (20, 8))
plt.rcParams['font.size'] = 18

plt.subplot(1, 2, 1)
sns.distplot(df_3['len'], color = 'blue', kde = False)
plt.title("H3F3A_bipolar_neuron.ENCFF068SXT.hg19.bed\n" + "Number of peaks =" + str(df_3.shape[0]))
plt.xlabel("Len")
plt.ylabel("Count")

plt.subplot(1, 2, 2)
sns.distplot(df_4['len'], color = 'red', kde = False)
plt.title("H3F3A_bipolar_neuron.ENCFF959PEX.hg19.bed\n" + "Number of peaks =" + str(df_4.shape[0]))
plt.xlabel("Len")
plt.ylabel("Count")

plt.savefig("hg19.histplot.png", format="png")

plt.show()
```

```python
df_3 = df_3[df_3['len'] < 4000]
df_4 = df_4[df_4['len'] < 1500]
```

```python
plt.figure(figsize = (20, 8))
plt.rcParams['font.size'] = 18

plt.subplot(1, 2, 1)
sns.distplot(df_3['len'], color = 'blue', kde = False)
plt.title("H3F3A_bipolar_neuron.ENCFF068SXT.hg19.filtered.bed\n" + "Number of peaks =" + str(df_3.shape[0]))
plt.xlabel("Len")
plt.ylabel("Count")

plt.subplot(1, 2, 2)
sns.distplot(df_4['len'], color = 'red', kde = False)
plt.title("H3F3A_bipolar_neuron.ENCFF959PEX.hg19.filtered.bed\n" + "Number of peaks =" + str(df_4.shape[0]))
plt.xlabel("Len")
plt.ylabel("Count")

plt.savefig("hg19.histplot.filtered.png", format="png")

plt.show()
```

```python
df_3.drop('len', axis = 1, inplace = True)
df_4.drop('len', axis = 1, inplace = True)
```

```python
df_3.to_csv('H3F3A_bipolar_neuron.ENCFF068SXT.hg19.filtered.bed', sep='\t', header=None, index=None)
df_4.to_csv('H3F3A_bipolar_neuron.ENCFF959PEX.hg19.filtered.bed', sep='\t', header=None, index=None)
```
